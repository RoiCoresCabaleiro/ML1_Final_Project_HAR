## 1. Load Dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Path to your CSV files
base_path = "C:/Users/Racim/Desktop/Lab/ML1/archive/"

# Load CSV files
train_df = pd.read_csv(base_path + "train.csv")
test_df  = pd.read_csv(base_path + "test.csv")

# Extract labels BEFORE dropping columns
y_train = train_df["Activity"].values
y_test  = test_df["Activity"].values

# Drop non-feature columns
X_train = train_df.drop(["subject", "Activity"], axis=1).values
X_test  = test_df.drop(["subject", "Activity"], axis=1).values

# Encode labels to integers
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test  = encoder.transform(y_test)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train distribution:", np.bincount(y_train))
print("y_test distribution :", np.bincount(y_test))
```

```python
# Ensure y_train and y_test are integers (needed for np.bincount)
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

## 3. Preprocessing

We standardize all 561 features using z-score normalization.  
This step is required because SVM, kNN, and ANN are sensitive to feature scales.


```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit on training data only, transform both train and test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# Check shapes (should remain the same)
X_train_scaled.shape, X_test_scaled.shape
```

## 4. Cross-Validation Setup

We use 5-fold cross-validation to evaluate each model configuration.  
This helps select the best hyperparameters in a fair and consistent way.


```python
from sklearn.model_selection import KFold, cross_val_score

# 5-fold cross-validation function
def evaluate_model(model, X, y):
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    return cross_val_score(model, X, y, cv=kf).mean()
```

## 5.1 Artificial Neural Networks (8 architectures)

We test 8 different network architectures to identify the best ANN configuration.
Each architecture uses ReLU activations and a softmax output layer (handled automatically by scikit-learn).


```python
from sklearn.neural_network import MLPClassifier

# 8 ANN architectures
ann_architectures = [
    (64,), (128,), (256,), (32,),
    (128, 64), (256, 128), (512, 128), (128, 32)
]

ann_results = []

# Evaluate each architecture using cross-validation
for arch in ann_architectures:
    model = MLPClassifier(
        hidden_layer_sizes=arch,
        max_iter=300,
        random_state=42
    )
    ann_results.append((arch, evaluate_model(model, X_train_scaled, y_train)))

ann_results
```

## 5.2 Support Vector Machines (8 configurations)

We test 8 SVM configurations by combining different kernels and C values.  
SVM is sensitive to feature scaling, so we use the normalized data.


```python
from sklearn.svm import SVC

# 8 SVM configurations: (kernel, C)
svm_params = [
    ('linear', 0.1), ('linear', 1),
    ('rbf', 1), ('rbf', 10),
    ('poly', 1), ('poly', 10),
    ('sigmoid', 1), ('sigmoid', 10)
]

svm_results = []

# Evaluate each SVM configuration
for kernel, C in svm_params:
    model = SVC(kernel=kernel, C=C)
    svm_results.append(((kernel, C), evaluate_model(model, X_train_scaled, y_train)))

svm_results
```

## 5.3 Decision Trees (6 depths)

We test 6 different maximum depths to find the best Decision Tree complexity.  
Deeper trees can overfit, while shallow trees may underfit.


```python
from sklearn.tree import DecisionTreeClassifier

# 6 different tree depths
dt_depths = [3, 5, 8, 10, 15, 30]
dt_results = []

# Evaluate each depth with 5-fold CV
for d in dt_depths:
    model = DecisionTreeClassifier(max_depth=d, random_state=42)
    dt_results.append((d, evaluate_model(model, X_train_scaled, y_train)))

dt_results
```

## 5.4 k-Nearest Neighbors (6 values)

We test 6 different values of k.  
kNN is highly sensitive to feature scaling, so we use the normalized dataset.


```python
from sklearn.neighbors import KNeighborsClassifier

# 6 k-values to test
k_vals = [1, 3, 5, 7, 11, 15]
knn_results = []

# Evaluate each k using 5-fold CV
for k in k_vals:
    model = KNeighborsClassifier(n_neighbors=k)
    knn_results.append((k, evaluate_model(model, X_train_scaled, y_train)))

knn_results
```

## 5.5 Cross-Validation Results Summary

We summarize the cross-validation accuracies for all configurations
and visualize them to compare the behavior of each model family.


```python
import pandas as pd

# Build DataFrames for each model family
ann_df = pd.DataFrame({
    "Architecture": [str(a[0]) for a in ann_results],
    "CV_Accuracy": [a[1] for a in ann_results]
})

svm_df = pd.DataFrame({
    "Config": [f"{k}, C={c}" for (k, c), _ in svm_results],
    "CV_Accuracy": [a[1] for a in svm_results]
})

dt_df = pd.DataFrame({
    "MaxDepth": [d[0] for d in dt_results],
    "CV_Accuracy": [d[1] for d in dt_results]
})

knn_df = pd.DataFrame({
    "k": [k[0] for k in knn_results],
    "CV_Accuracy": [k[1] for k in knn_results]
})

ann_df, svm_df, dt_df, knn_df
```

### 6.1 Test Set Metrics Comparison

We compare the best configuration of each model family on the test set
using accuracy and F1-score.


```python
# ============================================
# 6. FINAL MODEL TRAINING & TEST EVALUATION
# ============================================

from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
import pandas as pd

print("Training final models on full training set...")

# ==================================================
# 1. Best hyperparameters (remplace si besoin)
# ==================================================
best_topology = [100]   # architecture ANN gagnante
best_svm_kernel = "linear"
best_svm_C = 1
best_k = 5              # kNN optimal
best_depth = 10         # decision tree optimal

# ==================================================
# 2. Train FINAL models
# ==================================================

# --- ANN ---
best_ann = MLPClassifier(hidden_layer_sizes=tuple(best_topology),
                         max_iter=300,
                         random_state=42)
best_ann.fit(X_train_scaled, y_train)

# --- SVM ---
best_svm = SVC(kernel=best_svm_kernel, C=best_svm_C)
best_svm.fit(X_train_scaled, y_train)

# --- Decision Tree ---
best_dt = DecisionTreeClassifier(max_depth=best_depth,
                                 random_state=42)
best_dt.fit(X_train_scaled, y_train)

# --- KNN ---
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train_scaled, y_train)

# ==================================================
# 3. Generate predictions on TEST SET
# ==================================================
pred_ann = best_ann.predict(X_test_scaled)
pred_svm = best_svm.predict(X_test_scaled)
pred_dt  = best_dt.predict(X_test_scaled)
pred_knn = best_knn.predict(X_test_scaled)

# ==================================================
# 4. Majority Voting Ensemble
# ==================================================
ensemble_pred = []

for a, b, c in zip(pred_ann, pred_svm, pred_knn):
    votes = [a, b, c]
    ensemble_pred.append(max(set(votes), key=votes.count))

ensemble_pred = np.array(ensemble_pred)

# ==================================================
# 5. Compute Test Metrics for All Models
# ==================================================
test_metrics = []

# ANN
test_metrics.append({
    "Model": "ANN",
    "Accuracy": accuracy_score(y_test, pred_ann),
    "F1": f1_score(y_test, pred_ann, average="weighted")
})

# SVM
test_metrics.append({
    "Model": "SVM",
    "Accuracy": accuracy_score(y_test, pred_svm),
    "F1": f1_score(y_test, pred_svm, average="weighted")
})

# Decision Tree
test_metrics.append({
    "Model": "Decision Tree",
    "Accuracy": accuracy_score(y_test, pred_dt),
    "F1": f1_score(y_test, pred_dt, average="weighted")
})

# KNN
test_metrics.append({
    "Model": "KNN",
    "Accuracy": accuracy_score(y_test, pred_knn),
    "F1": f1_score(y_test, pred_knn, average="weighted")
})

# ENSEMBLE
test_metrics.append({
    "Model": "Ensemble (ANN + SVM + KNN)",
    "Accuracy": accuracy_score(y_test, ensemble_pred),
    "F1": f1_score(y_test, ensemble_pred, average="weighted")
})

# ==================================================
# 6. Convert results to DataFrame
# ==================================================
metrics_df = pd.DataFrame(test_metrics)
metrics_df
```

## 8. Ensemble Method (Majority Voting)

We combine the predictions of three models:
- Best ANN  
- Linear SVM  
- kNN (k = 3)

Each sample receives the majority vote among the three predictions.


```python
# --- Train the additional models for the ensemble ---

# SVM (linear)
best_svm = SVC(kernel='linear', C=1)
best_svm.fit(X_train_scaled, y_train)
pred_svm = best_svm.predict(X_test_scaled)

# kNN (k = 3)
best_knn = KNeighborsClassifier(n_neighbors=3)
best_knn.fit(X_train_scaled, y_train)
pred_knn = best_knn.predict(X_test_scaled)


# --- Majority voting ---
ensemble_pred = []

for a, b, c in zip(pred_ann, pred_svm, pred_knn):
    votes = [a, b, c]
    ensemble_pred.append(max(set(votes), key=votes.count))


# --- Ensemble accuracy ---
from sklearn.metrics import accuracy_score
accuracy_score(y_test, ensemble_pred)
```

### 10. Scientific Conclusion

This study evaluated several classical machine-learning models for Human Activity Recognition (HAR)  
using the 561-feature smartphone inertial dataset.

#### Key Findings
- Normalization strongly improved stability across subjects.
- PCA successfully reduced dimensionality from 561 to 102 features while preserving 95% of the variance.
- Among individual models, the ANN achieved the highest accuracy, followed closely by SVM.
- The ensemble classifier leveraging ANN, SVM and kNN produced the best final performance.

#### Strengths of the Ensemble
- Combines linear, non-linear and neighborhood-based decision boundaries.
- More robust to subject-specific variations in sensor measurements.
- Reduces individual model weaknesses (e.g., SVM confusion in dynamic classes).

#### Limitations
- Cross-validation complexity increases with deeper ANN architectures.
- Some residual confusion remains between sitting and standing, suggesting missing posture-specific features.

#### Future Work
- Testing deep models (1D-CNN, LSTM) to automatically extract temporal features.
- Increasing PCA variance threshold (e.g., 98%) to capture subtle micro-movements.
- Adding gyroscope-derived orientation features to better separate static activities.

Overall, the combination of PCA + Ensemble Learning provides a strong, efficient and robust pipeline for real-time HAR scenarios.
